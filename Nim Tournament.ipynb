```python
import os
import importlib
import sys
```

```python
def get_players(p1, p2):
    '''
    Given the name of two students in the directory this function returns
    imported NimPlayer instantiations. 

    input:
        p1: string
        p2: string
    
    output:
        player1: NimPlayer()
        player2: NimPlayer()
    '''
    sys.path.append(p1)
    player1 = __import__(p1, fromlist=['nim_player']).nim_player.NimPlayer()
    sys.path.append(p2)
    player2 = __import__(p2, fromlist=['nim_player']).nim_player.NimPlayer()

    return player1, player2
```

```python
def play_game(player1, player2):
    '''
    Given two nim_player classes this function will play a 
    game and return True if player1 wins or False if player2 wins.
    input:
        player1: nim_player class
        player2: nim_player class
    output:
        Boolean
    '''
    state = [1,3,5,7]
    ended = False
    while(not ended):
        state = player1.play(state)

        if sum(state)==0:
            #print('player1 lost')
            return False
        state = player2.play(state)

        if sum(state)==0:
            #print('player2 lost')
            return True
```

```python
def tournament(players):
    score_keeper = {}

    for j in range(len(players)):
        score_keeper[players[j]] = [0,0]
    for j in range(len(players)):
        temp_players = players.copy()
        
        temp_players.pop(j)
        for i in range(len(temp_players)):
            try:
                player1, player2 = get_players(players[j],temp_players[i])
            except:
                print('not working', temp_players[i], players[j])
                continue
            
            
            if play_game(player1,player2):
                score_keeper[players[j]][0] += 1
                score_keeper[temp_players[i]][1] += 1
            
            else:
                score_keeper[players[j]][1] += 1
                score_keeper[temp_players[i]][0] += 1
    return score_keeper
```

```python
def show_game(state,p):
    print('Player '+p)
    big = state[-1]
    for x in state:
        bars = ''
        for z in range(int((big-x)/2)):
            bars += ' '
        for y in range(x):
            bars += '|'
        print(bars)
    print('\n')
    
```

```python
def tournament_show(players):
    score_keeper = {}

    for j in range(len(players)):
        score_keeper[players[j]] = [0,0]
    for j in range(len(players)):
        temp_players = players.copy()
        temp_players.pop(j)
        for i in range(len(temp_players)):
            try:
                player1, player2 = get_players(players[j],temp_players[i])
                print(players[j], 'vs', temp_players[i])
            except:
                print('not working', temp_players[i], players[j])
                continue
            
            
            if play_game_show(player1,player2):
                score_keeper[players[j]][0] += 1
                score_keeper[temp_players[i]][1] += 1
            
            else:
                score_keeper[players[j]][1] += 1
                score_keeper[temp_players[i]][0] += 1
    return score_keeper
```

```python
def play_game_show(player1, player2):
    '''
    Given two nim_player classes this function will play a 
    game and return True if player1 wins or False if player2 wins.
    input:
        player1: nim_player class
        player2: nim_player class
    output:
        Boolean
    '''
    state = [1,3,5,7]
    ended = False
    show_game(state, '0')
    while(not ended):
        state = player1.play(state)
        show_game(state,'1')
        if sum(state)==0:
            #print('player1 lost')
            return False
        state = player2.play(state)
        show_game(state,'2')
        if sum(state)==0:
            #print('player2 lost')
            return True
```
